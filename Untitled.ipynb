```python
results = {}
```

```python
users = list(set(df['name']))
for u in users:
    results[u] = {}
    udf = df[df['name'] == u]
    
    results[u]['Сообщений'] = len(udf)
    results[u]['Медиа'] = len(udf[udf['is_media'] == True])
    results[u]['Фоточек'] = len(udf[udf['is_photo'] == True])
    results[u]['Стикеров'] = len(udf[udf['is_sticker'] == True])
    results[u]['Анимаций'] = len(udf[udf['is_animation'] == True])
    results[u]['Репосты'] = len(udf[udf['repost'] == True])
    results[u]['Реплаи'] = len(udf[udf['reply'] > 0])
    results[u]['Голосовалки'] = len(udf[udf['poll'] == True])
```

```python
itog = pd.DataFrame.from_dict(results)
summ = pd.DataFrame(itog.T.sum(), columns=['ИТОГО'])

pd.concat([itog, summ], axis=1).T
```
